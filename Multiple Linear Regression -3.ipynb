```python
map_val={'TA':2,'Gd':3,'Fa':1,'Ex':4,'Po':1,'None':0,'Y':1,'N':0,'Reg':3,'IR1':2,'IR2':1,'IR3':0,'None':0,'No':2,'Mn':2,'Av':3,
        'Gd':4,'Unf':1,'LwQ':2,'Rec':3,'BLQ':4,'ALQ':5,'GLQ':6}
```

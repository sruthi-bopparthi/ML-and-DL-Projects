### **Step 2: Import Beautiful Soup**

* Once installed, you need to import Beautiful Soup and possibly another library like requests to fetch web pages.
* Here's how you can import them:

```python
import requests
from bs4 import BeautifulSoup
```

## **Step 3: Fetch a Web Page**

* To scrape data from a web page, you first need to fetch its HTML content.
* You can use the requests library for this:

```python
url = "https://www.worldometers.info/gdp/gdp-per-capita/"  # Replace with the URL of the webpage you want to scrape
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html = response.text
    print(html)
else:
    print("Failed to retrieve the webpage.")
```

## **Step 4: Create a Beautiful Soup Object**
* Next, you create a Beautiful Soup object by passing the HTML content and specifying the parser to use (usually "html.parser"):

```python
soup = BeautifulSoup(html, 'html.parser')
```

### **Step 5: Navigate and Extract Data**
* Now, you can navigate the HTML document and extract the data you need.
* Beautiful Soup provides various methods and attributes to help you locate elements.
* Here are some common ones:

 * find(): Find the first occurrence of a specified tag.
 * find_all(): Find all occurrences of a specified tag.
 * select(): Use CSS selectors to locate elements.

```python
# Find all the <a> tags
links = soup.find_all('a')

# Print the text and href attribute of each link
for link in links:
    print("Text:", link.text)
    print("URL:", link.get('href'))
    print()
```

**Example 2: Scraping a Table**
* Suppose you want to scrape a table from a webpage.
 * First, find the table by inspecting the HTML source code or using developer tools in your browser.
 * Once you've identified the table, you can extract its data:



 image.png

```python
# Find the table (replace 'table_id' with the actual table's ID or class)

# table = soup.find('table', id='table_id')

# How to get table id
# Go to the table in the website
# Right click on table and select inspect get the ID

table = soup.find('table', id='example2')

# Extract data from the table rows and columns
for row in table.find_all('tr'):
    columns = row.find_all('td')
    for column in columns:
        print(column.text, end='\t')
    print()
```

### **Step 6: Data Processing and Storage**

Let see how we can put the data into a dataframe

```python
#Import pandas
import pandas as pd
```

```python
# Let modify our codes above
# Initialize empty lists to store the table data
data = []

# Extract data from the table rows and columns
for row in table.find_all('tr'):
    columns = row.find_all('td')
    row_data = [column.text for column in columns]
    data.append(row_data)
```

Once you have the data in lists, you can create a Pandas DataFrame from it:

```python
# Create a DataFrame from the data
data_GDP = pd.DataFrame(data)

data_GDP.head()
```

* Let us add column headers

```python
# Let add column heads
column =['Remove me','Country','GDP (PPP) per capita (2022)', 'GDP (nominal) per capita (2022)' , 'vs. World PPP GDP per capita ($20, 645)']

# Create a DataFrame from the data
data_GDP = pd.DataFrame(data, columns=column )

data_GDP.head()
```

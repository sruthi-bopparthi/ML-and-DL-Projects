# Multi-Index and Index Hierarchy
Let us go over hoe to work with Multi index, first we will create a quick exmaple of what a multi-indexed Data fram would look like:

```python
#index levels
outside=['North','North','North','South','South','South']
inside=newind
```

```python
hier_index=list(zip(outside,inside))
```

```python
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df=pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
```

```python
df.loc['North']
```

```python
df.loc['North'].loc['UP']
```

```python
df.index.names=['Region','States']
```

```python
df
```

```python
#return cross-section from the dataframseries
df.xs('North')
```

```python
df.xs(['North','DEL'])
```
